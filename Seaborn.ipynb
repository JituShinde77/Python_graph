```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.scatterplot(x=df.sepal_length, y=df.petal_length)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.sepal_length, df.petal_width)
plt.xlabel('sep_len')
plt.ylabel('plt_len')
```

```python
plt.plot(df.sepal_length, df.petal_width, 'o')
plt.xlabel('sep_len')
plt.ylabel('plt_len')
```

```python
plt.hist(df.sepal_length, alpha= .5)
plt.xlabel('S_len')
plt.ylabel('Freq')
```
